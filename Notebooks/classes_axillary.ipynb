```python
import time
from enum import Enum

class EvaluationFocus(Enum):

    drift_types = 'drift_type'
    change_types = 'change_type'


class DriftType(Enum):

    default = 'other'
    simple = 'simple'
    simple_sudden = 'sudden'
    simple_gradual = 'gradual'
    complex_recurring = 'recurring'
    complex_incremental = 'incremental'


class ChangeType(Enum):
    sudden = 'sudden'
    gradual_start = 'gradual_start'
    gradual_end = 'gradual_end'
    gradual = 'gradual'

class RelationChangeType(Enum):
    sudden = 'sudden'
    gradual = 'gradual'
    none = 'NA'


class ColumnNames(Enum):
    percent = '_%'
    change_rate_short = '->rate->'
    trend = '->trend->'
    weight = 'weight'
    cumsum_weight = 'cumsum_weight'
    change_type = 'change_type'
    noise = 'noise'
    change_rate = 'change_rate'

class TrendTypes(Enum):
    constant = 'constant'
    noise = 'noise'
    increase = 'increase'
    decrease = 'decrease'
    unknown = 'unknown'

class ChangeMomentsInputFile(Enum):
    complexity = 'complexity'
    log_name = 'log_name'
    drift_id = 'drift_id'
    drift_type = 'drift_type'
    change_id = 'change_id'
    change_type = 'change_type'
    change_index = 'change_index'
    timestamp = 'timestamp'
    change_start_end = 'change_start_end'

class EventAttributeNames(Enum):
    concept_name = 'concept:name'
    timestamp = 'time:timestamp'
    start_event = 'start_event'
    end_event_name = 'end_event'

class ResultsData(Enum):
    log_name = 'log_name'
    change_id = 'change_id'
    change_moment = 'change_moment'
    change_type = 'change_type'
    drift_type = 'drift_type'
    change_index = 'change_index'
    window_before = 'window_before'
    window_after = 'window_after'


class ExecutionTiming:

    def __init__(self):

        self.main_tic = time.perf_counter()
        self.main_tic_timestamp = time.localtime()
        self.interim_tic = time.perf_counter()
        self.interim_toc = time.perf_counter()
        print('-' * 80)
        print(f"Start script execution: {time.asctime(self.main_tic_timestamp)}")
        print('-' * 80)
        print()

    def report_interim(self):
        self.interim_toc = time.perf_counter()
        self.run_time = self.interim_toc - self.interim_tic
        #print('-' * 80)
        print(f"Interim execution time: {self.run_time:0.1f} seconds, {self.run_time / 60:0.1f} minutes, "
              f"{self.run_time / 60 / 60:0.1f} hours")
        print('-' * 80)
        self.interim_tic = time.perf_counter()
        return None

    def report_end(self):
        self.main_toc_timestamp = time.localtime()
        self.main_toc = time.perf_counter()
        self.run_time = self.main_toc - self.main_tic

        print()
        print()
        print('-' * 80)
        print("Successful program execution")
        print(f"Program start: {time.asctime(self.main_tic_timestamp)}")
        print(f"Program end: {time.asctime(self.main_toc_timestamp)}")
        print(f"Program finished in {self.run_time:0.1f} seconds, {self.run_time / 60:0.1f} minutes,"
              f"{self.run_time / 60 / 60:0.1f} hours")
        print('-' * 80)
        return None
```
